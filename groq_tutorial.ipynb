```python
import os
import shutil
import site

paths = site.getsitepackages()

targets = [
    "~rllib3",
    "~rllib3.dist-info",
    "urllib3",
    "urllib3.dist-info"
]

for base in paths:
    for t in targets:
        p = os.path.join(base, t)
        if os.path.exists(p):
            shutil.rmtree(p, ignore_errors=True)
            print(f"Removed: {p}")
```
